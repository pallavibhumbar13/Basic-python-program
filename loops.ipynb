```python
#stop loop at 5
```

```python
for i in range(1,11):

    if i == 5:
        break

    print(i)
```

```python
for i in range(11,20):

    if i == 16:
        break
    print(i)
```

```python
for i in range(11,20):

    if i == 16:
        continue
    print(i)
```
