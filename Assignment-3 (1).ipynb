```python
# 13.Write a Python program to calculate electricity bill charges based on unit slab rates and display the final bill amount.
```

```python
# Calculate electricity bill chargers based on unit slab rates and display final bill amount
```

```python
units = int(input("Enter electricity units consumed: "))
bill = 0
if units <= 100:
    bill = units * 1.5
elif units <= 200:
    bill = (100 * 1.5) + (units - 100) * 2.5
elif units <= 300:
    bill = (100 * 1.5) + (100 * 2.5) + (units - 200) * 4
else:
    bill = (100 * 1.5) + (100 * 2.5) + (100 * 4) + (units - 300) * 6
print("Total Electricity Bill = ₹", bill)
```

```python
choice = int(input("Enter your choice (1-4): "))
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
if choice == 1:
    print("Result =", a + b)
elif choice == 2:
    print("Result =", a - b)
elif choice == 3:
    print("Result =", a * b)
elif choice == 4:
    if b != 0:
        print("Result =", a / b)
    else:
        print("Cannot divide by zero")
else:
    print("Invalid choice")
```

```python
a = int(input("Angle 1: "))
b = int(input("Angle 2: "))
c = int(input("Angle 3: "))
if a + b + c == 180:
    print("Valid Triangle")
else:
    print("Not a Valid Triangle")
```

```python
# Identify whether it is acute, right, or obtuse
```

```python
if a + b + c == 180:
    if a == 90 or b == 90 or c == 90:
        print("Right Triangle")
    elif a > 90 or b > 90 or c > 90:
        print("Obtuse Triangle")
    else:
        print("Acute Triangle")
else:
    print("Not a Valid Triangle")
```

```python
price = float(input("Enter product price: "))
print("Product price is:", price)
```

```python
# Apply discount based on price range and calculate final price
```

```python
if price <= 1000:
    discount = 0
elif price <= 5000:
    discount = price * 0.05
elif price <= 10000:
    discount = price * 0.10
else:
    discount = price * 0.15
final_price = price - discount

print("Discount =", discount)
print("Final Price =", final_price)
```
