```python
#creates sample file
```

```python
with open("sample.txt","w") as f:
    f.write("name:pallavi\n")
    f.write("role:student\n")
    f.write("location:india\n")
```

```python
#reading the file
```

```python
with open("sample.txt","r")as f:
    data = f.read()
    print(data)
```

```python
#reading line byline
```

```python
with open("sample.txt","r") as f:
    for line in f:
        print(line.strip())
```
