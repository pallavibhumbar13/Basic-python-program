```python
try:
    num=int("abc")
    result = 10/num
except ValueError:
    print("conversion failed")
except ZeroDivisionError:
    print("cannot divide by zero")
```

```python
try:
    x=10/0
    #result = 10/num
except ValueError:
    print("conversion failed")
except ZeroDivisionError:
    print("cannot divide by zero")
```

```python
#multiple exception in one except
# you can handle more than one error type in a single block.
```

```python
try:
    x = int("abc")
except(ValueError,TypeError):
    print("Value error or type error occured")
```

```python
# generic exception
#catches any kind of error when you don't specify the exact type
```

```python
try:
    x=int("abc")
except Exception as e:
    print("error occured:",e)
```

```python
try:
    x=10/0
except Exception as e:
    print("error occured:",e)
```

```python
#else block
#runs only when no error occurs in the try block.
```

```python
try:
    x = 10 /2
except ZeroDivisionError:
    print("error")
else:
    print("success:",x)
```

```python
#finally block
#runs every time,whether an errors happens or not,often used for cleanup.
```

```python
try:
    x = 10 / 0
except ZeroDivisionError:
    print("error")
finally:
    print("this always runs")
```
