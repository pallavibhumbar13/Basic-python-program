```python
#6. Write a Python program to calculate the factorial of a number, count its digits, and display the reverse of the number.
```

```python

num = int(input("Enter a number: "))

factorial = 1
for i in range(1, num + 1):
    factorial *= i

# Count digits
digit_count = len(str(abs(num)))

# Reverse the number
reverse_num = str(abs(num))[::-1]

# Display results
print("Factorial =", factorial)
print("Number of digits =", digit_count)
print("Reverse of the number =", reverse_num)
```

```python
#11. Write a Python program to reverse a string, count the vowels, and count the consonants.
```

```python
# Input a string
s = input("Enter a string: ")

# Reverse the string
reverse_s = s[::-1]

# Count vowels and consonants
vowels = "aeiouAEIOU"
vowel_count = 0
consonant_count = 0

for ch in s:
    if ch.isalpha():  # Check if character is a letter
        if ch in vowels:
            vowel_count += 1
        else:
            consonant_count += 1

# Display results
print("Original String:", s)
print("Reversed String:", reverse_s)
print("Number of Vowels:", vowel_count)
print("Number of Consonants:", consonant_count)
```
