```python
# **keyword argument  (key value)
```

```python
def info(**data):
    print(data)

info(name="pallavi",age=17, marks=90, clas="tyit")
```

```python
# * args  (multiple values)
```

```python
def add(*nums):
    print(sum(nums))

add(10,20,30)
```

```python
def mul(*nums):
    i = 1
    for num in nums:
        i *= num
    print(i)

mul(10, 20, 30)  
```

```python
#lambda functions   (Anonymous function)
```

```python
add = lambda a,b: a+b

print(add(10,7))
```

```python
add = lambda a,b: a*b

print(add(2,5))
```

```python
numbers=[1,2,3,4,5,6,8]
```

```python
numbers
```

```python
for num in numbers:
    if num % 2 == 0:
        print(f"{num} is even")
    else:
        print(f"{num} is odd")
```

```python
i = 0

for num in numbers:
    if num % 2 == 0:
        i += 1

print(f"Total: {i}")
```
